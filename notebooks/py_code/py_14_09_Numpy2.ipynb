```python
import numpy as np
```

```python
# a simple toss coin simulation

coin = ["heads", "tails"]
number_of_throws = 10000
throws = np.random.choice(coin, number_of_throws)
```

```python
np.sum(throws=="heads")/number_of_throws
```

```python
np.sum(throws=="tails")/number_of_throws
```

```python
probablities = [0.3, 0.7]

throws_with_p = np.random.choice(coin, number_of_throws, p=probablities)
print(np.sum(throws_with_p=="heads"))
print(np.sum(throws_with_p=="tails"))
```

# A coin landing on a side

```python
coin_toss_side = ["heads", "tails", "side"]
throws = 100000

throws_100000 = np.random.choice(coin_toss_side, size=throws, p=[0.4995, 0.4995, 0.001])

total_heads_side = np.sum(throws_100000 == "heads")
total_tails_side = np.sum(throws_100000 == "tails")
total_side = np.sum(throws_100000 == "side")

print("Total heads:", total_heads_side)
print("Total tails:", total_tails_side)
print("Total side:", total_side)
```

```python
unique, counts = np.unique(throws_100000, return_counts=True)
print(counts)

import matplotlib.pyplot as plt
plt.bar(unique, counts)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.bar(coin_toss_side, [total_heads_side, total_tails_side, total_side], color="thistle")
plt.title("Coin Toss with Side landing")
plt.xlabel("Landing result")
plt.ylabel("Frequency")
plt.show()
```
