#BROADCASTING

Numpy arrays can be broadcasted allowing to work with arrays of different shapes.

NumPy expands smaller shape to match the larger shape.

```python
# a matrix of student scores

# 3 students, 4 subjects
scores = np.array([
    [60, 54, 68, 91],
    [75, 87, 60, 83],
    [75, 74, 81, 97]
])
```

```python
scores.shape
```

```python
bonuses = np.array([5, 3, 10, 7])
bonuses.shape # one-dimensional array
```

```python
adjusted_scores = scores + bonuses
adjusted_scores
```

```python
bonuses2 = [10, 3, 7, 0]
adjusted_scores2 = scores + bonuses2
adjusted_scores2 = [score + bonus for score, bonus in zip(scores, bonuses2)]
print(adjusted_scores2)
```

#Random walk

```python
# a mathematical concept

# 2-dimensional, up right left down

import random

n = 100000 # number of steps
x = np.zeros(n)
y = np.zeros(n)

val = random.randint(1,4) # since we have 4 directions

for i in range(1,n):
  val = random.randint(1,4) # new random value
  # 1 = up, 2 = down, 3 = left, 4 = right
  if val == 1:
    x[i] = x[i-1] + 1 # dot moves: x coordinate +1
    y[i] = y[i-1] # y coordinate stays the same
  elif val == 2:
    x[i] = x[i-1] - 1 # dot moves: x coordinate -1
    y[i] = y[i-1] # y coordinate stays the same
  elif val == 3:
    x[i] = x[i-1] # x coordinate stays the same
    y[i] = y[i-1] + 1 # dot moves: y coordinate +1
  else:
    x[i] = x[i-1] # x coordinate stays the same
    y[i] = y[i-1] - 1 # dot moves: y coordinate -1
```

```python
  import pylab as pl
  pl.plot(x,y, color = "thistle")
  plt.title("Random walk, two-dimensional")
  plt.xlabel("x coordinate")
  plt.ylabel("y coordinate")
  plt.grid(True)
  pl.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))  # Increase figure size
plt.plot(x, y, color='blue', lw=2, alpha=0.7)  # Line with transparency
plt.scatter(x[0], y[0], c='green', s=100, label='Start')  # Mark start point
plt.scatter(x[-1], y[-1], c='red', s=100, label='End')  # Mark end point

plt.title('2D Random Walk', fontsize=16)
plt.xlabel('X-Coordinate')
plt.ylabel('Y-Coordinate')
plt.grid(True)
plt.legend()

plt.show()
```

```python
# TASK
# Simulate a 1D simple random walk where at each step can be moving only up or down. How many steps   can you actually get?

import numpy as np
import matplotlib.pyplot as plt

n = 1000 #steps

x = np.zeros(n)
y = np.arange(n)

# Simulation
for i in range(1, n):
    if np.random.rand() < 0.5:
        x[i] = x[i - 1] + 1
    else:
        x[i] = x[i - 1] - 1

# Colormap
colors = plt.cm.rainbow(np.linspace(0, 1, n))

# Plotting - with rainbow colors
for i in range(1, n):
    plt.plot(x[i-1:i+1], y[i-1:i+1], color=colors[i])

plt.xlabel("Position")
plt.ylabel("Step")
plt.title("One-dimensional random walk")
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
import numpy as np

steps = 1000

position = [0]

for i in range(steps):
    step = random.choice([-1, 1])
    position.append(position[-1] + step)

# rainbow colormap
colors = plt.cm.rainbow(np.linspace(0, 1, steps + 1))

plt.figure(figsize=(10, 5))
plt.scatter(range(steps + 1), position, c=colors)

plt.title("One-dimensional random walk")
plt.xlabel("Steps")
plt.ylabel("Position")
plt.grid(True)

plt.show()
```
