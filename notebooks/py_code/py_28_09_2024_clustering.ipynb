### Unsupervised training example (clustering)
https://scikit-learn.org/dev/api/sklearn.datasets.html

```python
from sklearn.datasets import make_blobs #creating synthetic dataset
```

```python
X, y = make_blobs(n_samples=100000, centers=10, cluster_std=0.80)
```

```python
X[:, 0] #x coordniate
X[:, 1] #y coordinate
y # the cluster of the point
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c='grey', edgecolors='k')
```

```python
from sklearn.cluster import KMeans #import model
```

```python
kmeans = KMeans(n_clusters=5) #initialize the model
```

```python
kmeans.fit(X) #train model
```

```python
cluster_labels = kmeans.predict(X) #predict
```

```python
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, edgecolors='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, marker='*', c='red')
```
