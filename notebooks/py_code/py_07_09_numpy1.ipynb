```python
import numpy as np
```

```python
coin = ["heads", "tails"] # one dimensional array - a list
```

```python
number_of_throws = 1000 # size of the random list we want to generate
```

```python
results = np.random.choice(coin, size=number_of_throws)
print(f"The results of {number_of_throws} coin tosses are: {results}")
```

```python
# how many times heads nad how many times tails

# heads = np.count_nonzero(results == "heads")
heads = np.sum(results == "heads")
# tails = np.count_nonzero(results == "tails")
tails = np.sum(results == "tails")

print(f"Heads: {heads}, tails: {tails}")
```

```python
# probabilities

heads_probability = heads / number_of_throws
tails_probability = tails / number_of_throws

print(f"Heads probability: {heads_probability}, tails probability: {tails_probability}") #
```

Design a simulation where you toss two coins simultaneously. Record the possible outcomes (heads or tails for each coin) and run the simulation for multiple trials (e.g., 100, 500, or 1000 tosses). After collecting the data:

    Calculate the probabilities of each outcome (HH, HT, TH, TT).
    Visualize the results using a bar chart or pie chart to represent the frequencies and probabilities of each outcome.
    Analyze whether the results align with the expected theoretical probabilities.

```python
import numpy as np

def coin_toss_simulation(number_of_tosses):
    coin = ["H", "T"]

    coin1 = np.random.choice(coin, number_of_tosses, replace=True)
    coin2 = np.random.choice(coin, number_of_tosses, replace=True)

    # combine the results of tosses for 2 coins
    results = np.char.add(coin1, coin2)

    # calculate the count of tosses
    HH_count = np.sum(results == "HH")
    HT_count = np.sum(results == "HT")
    TH_count = np.sum(results == "TH")
    TT_count = np.sum(results == "TT")

    return HH_count, HT_count, TH_count, TT_count

#bar charts for visualisation
import matplotlib.pyplot as plt

def plot_results(tosses, counts, title):
    outcomes = ["HH", "HT", "TH", "TT"]
    plt.bar(outcomes, counts, color=['blue', 'orange', 'green', 'red'])
    plt.xlabel("Outcome")
    plt.ylabel("Frequency")
    plt.title(title)
    plt.show()

def plot_probabilities(tosses, counts, title):
    outcomes = ["HH", "HT", "TH", "TT"]
    probabilities = [count / tosses for count in counts]
    plt.bar(outcomes, probabilities, color=['blue', 'orange', 'green', 'red'])
    plt.xlabel("Outcome")
    plt.ylabel("Probability")
    plt.title(title)
    plt.show()

# for 100 tosses
HH_count_100, HT_count_100, TH_count_100, TT_count_100 = coin_toss_simulation(100)
print("Results for 100 tosses:")
print(f"HH: {HH_count_100}, HT: {HT_count_100}, TH: {TH_count_100}, TT: {TT_count_100}")
plot_results(100, [HH_count_100, HT_count_100, TH_count_100, TT_count_100], "Results for 100 tosses")
plot_probabilities(100, [HH_count_100, HT_count_100, TH_count_100, TT_count_100], "Probabilities for 100 tosses")

# for 500 tosses
HH_count_500, HT_count_500, TH_count_500, TT_count_500 = coin_toss_simulation(500)
print("Results for 500 tosses:")
print(f"HH: {HH_count_500}, HT: {HT_count_500}, TH: {TH_count_500}, TT: {TT_count_500}")
plot_results(500, [HH_count_500, HT_count_500, TH_count_500, TT_count_500], "Results for 500 tosses")
plot_probabilities(500, [HH_count_500, HT_count_500, TH_count_500, TT_count_500], "Probabilities for 500 tosses")

# for 1000 tosses
HH_count_1000, HT_count_1000, TH_count_1000, TT_count_1000 = coin_toss_simulation(1000)
print("Results for 1000 tosses:")
print(f"HH: {HH_count_1000}, HT: {HT_count_1000}, TH: {TH_count_1000}, TT: {TT_count_1000}")
plot_results(1000, [HH_count_1000, HT_count_1000, TH_count_1000, TT_count_1000], "Results for 1000 tosses")
plot_probabilities(1000, [HH_count_1000, HT_count_1000, TH_count_1000, TT_count_1000], "Probabilities for 1000 tosses")
```
